# Importing Packages

```python
import torch
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling
)
from datasets import load_dataset
```

# 🔹 2. Load and Preprocess the Dataset (emails.csv)

```python
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

# 🔹 3. Load GPT-2 Model and Tokenizer

```python
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
    model.resize_token_embeddings(len(tokenizer))
```

# Use GPU if Available

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

# 🔹 4. Prepare Dataset for Training

```python
dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="train[:1%]")

def tokenize_function(example):
    return tokenizer(example["text"], truncation=True, padding="max_length", max_length=128)

tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["text"])
```

# 🔹 5. Define Training Configuration  

```python
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)
```

```python
print(torch.cuda.get_device_name(0))
print(torch.cuda.is_available())
```

# 🔹 6. Start Training

```python
training_args = TrainingArguments(
    output_dir="./trained_model",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    logging_steps=10,
    save_strategy="epoch",
    report_to="none",
    fp16=torch.cuda.is_available()
)
```

```python
# 8. Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer
)
```

```python
# 9. Train
trainer.train()
```

# 🔹 7. Generate a Sample Marketing Email

```python
model_path = "./trained_model"
model.save_pretrained(model_path)
tokenizer.save_pretrained(model_path)
```

```python
# 11. Generate email-style text
prompt = "Dear team, the training process has been successfully completed. "
inputs = tokenizer(prompt, return_tensors="pt").to(device)

with torch.no_grad():
    generated_ids = model.generate(
        inputs["input_ids"],
        max_length=100,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True
    )

generated_text = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
```

```python
# 12. Print output
print("\nGenerated Email:\n")
print(generated_text)
```
